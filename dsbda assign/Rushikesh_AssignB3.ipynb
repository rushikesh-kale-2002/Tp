###  Data Visualization - Pie Chart

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
file = pd.read_csv("C:\\Users\\HP\\Desktop\\airquality1.csv")
file
```

```python
file["Ozone"]= file["Ozone"].fillna(file["Ozone"].mean())
file["Solar.R"]= file["Solar.R"].fillna(file["Solar.R"].mean())
```

```python
file
```

```python
sizes = [file['Ozone'].mean(),file['Solar.R'].mean(),file['Wind'].mean(),file['Temp'].mean()]
colors = ['purple','greenyellow','blue','red']
labels = 'Ozone','Solar.R','Wind','Temp'
explode = (0.1,0,0,0)
```

```python
plt.pie(sizes,explode=explode, labeldistance=1.1,labels=labels, colors=colors)

plt.title("Average Data")
plt.savefig('plot.png')
```

```python
h = file.iloc[1:21,3]
h
```

### Bar graph 

```python
import numpy as np
y_posi=np.arange(len(h))
v=range(1,21)
```

```python
plt.bar(y_posi, height=h, width=0.8,color="red", bottom=None, data=None)
plt.xticks(y_posi,v)
plt.savefig('plot1.png')
plt.title('Temperature of 1st-20th May')
plt.xlabel('Days')
plt.ylabel('Temprature')
plt.show()
```
